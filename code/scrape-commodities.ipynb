```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
def scrape_commodities(monthyear,lock_id):
    url = 'https://corpslocks.usace.army.mil/lpwb/xml.tonnage?in_river=OH&in_lock=' + lock_id + '&in_mon_yr=' + monthyear
    page = requests.get(url, verify=False)

    # Create a BeautifulSoup object
    soup = BeautifulSoup(page.text, 'xml')
    
    rows = soup.findAll('ROW')

    for row in rows:
        comm_code = row.COMM_CODE.contents[0]
        comm_desc = row.COMM_DESC.contents[0]
        try:
            up_tons = row.UPBOUND_TONS.contents[0]
        except AttributeError:
            up_tons = 0
        try:
            down_tons = row.DOWNBOUND_TONS.contents[0]
        except AttributeError:
            down_tons = 0
        total_tons = row.TOTAL_TONS.contents[0]
        
        f.writerow([monthyear,lock_id,comm_code,comm_desc,up_tons,
                    down_tons,total_tons])
```

```python
f = csv.writer(open('../data/commodity/scraped-commodities.csv', 'w'))
f.writerow(['monthyear','lock_id','comm_code', 'comm_desc','up_tons',
            'down_tons','total_tons'])

monthyear_list = ['082018','092018','102018','112018','122018','012019',
                 '022019','032019','042019','052019','062019','072019',
                 '082019','092019','102019']#,'112019','122019','012019']
lock_list = ['01','02','03','04','05','21','22','24','25','26','41','42',
             '71','72','75','76','77','78','79']

for monthyear in monthyear_list:
    for lock_id in lock_list:
        scrape_commodities(monthyear,lock_id)
```
